```python
s_input_2 = """Player 1:
43
19

Player 2:
2
29
14"""

puz_input = s_input_2.split("\n")


input_file="data/day22.txt"
puz_input = [ x.strip() for x in open(input_file).readlines()]


PART1 = False
players={}
p=""
for x in puz_input:
    if x == "":
        continue
    if  x[0] == "P":
        p = x.strip().split(" ")[1][0:-1]
        players[p] = []
    else:
        players[p].append(int(x))



def game(deck1, deck2):
    prev_hands = []
    while len(deck1) > 0  and len(deck2) > 0:
        
        s = "{}_{}".format( ",".join([str(x) for x in deck1]) , ",".join([str(x) for x in deck2]) )
        if  s in prev_hands:
            return [deck1, []]
        else:
            prev_hands.append( s )
            
            
        p1card = deck1.pop(0)
        p2card = deck2.pop(0)

        if len(deck1) >= p1card and len(deck2) >= p2card:
            [d1,d2] = game(deck1[0:p1card], deck2[0:p2card])
            if len(d1) == 0:
                deck2.append(p2card)
                deck2.append(p1card)
            else:
                deck1.append(p1card)
                deck1.append(p2card)
        else:        
            if p1card > p2card:
                deck1.append(p1card)
                deck1.append(p2card)
            else:
                deck2.append(p2card)
                deck2.append(p1card)

        
        
    return [deck1,deck2]

print(game( players['1'], players['2']))
print(sum([ (idx+1)*v for idx,v in enumerate(reversed(players['1']))]))
```
