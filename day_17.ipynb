```python
import copy

s_input_1 = """.#.
..#
###"""

input_1 = """.#######
#######.
###.###.
#....###
.#..##..
#.#.###.
###..###
.#.#.##."""

# puz_input = s_input_1
puz_input = input_1

data = {}
for yidx,yval in enumerate(puz_input.split("\n")):
    for xidx, xval in enumerate( list(yval) ):
        data ["{}_{}_{}".format(xidx, yidx, 0)] = xval
        
# print(data)

## Cycles through known points, if a points doesn't exist, create it as inactive .

def count_neighbors(data, x, y, z):
    t = 0
    for _z in [-1,0,1]:
        for _x in [-1,0,1]:
            for _y in [-1,0,1]:
                if _z == 0 and _x == 0 and _y == 0:
                    continue
                p = "{}_{}_{}".format(x+_x, y+_y, z+_z)
                if p in data and data[p] == "#":
                    t +=1
    return t


def create_new_data(data):
    new_data = {}
    for k in data:
        [x,y,z] = [int(x) for x in k.split("_")]
        for _z in [-1,0,1]:
            for _x in [-1,0,1]:
                for _y in [-1,0,1]:
                    p = "{}_{}_{}".format(x+_x, y+_y, z+_z)
                    if p in data:
                        new_data[p] = data[p]
                    else:
                        new_data[p] = "."
    return new_data

for i in range(0,6):               
    # add all the next space        
    data = create_new_data(data)

    new_data = {}
    # look to see what changes
    for k in data:
        [x,y,z] = [int(x) for x in k.split("_")]
        p = "{}_{}_{}".format(x, y, z)
        v = count_neighbors(data,x,y,z)
        if data[k] == "#" and v in [2,3]:
            new_data[p] = "#"
        elif data[k] == "#" and v not in [2,3]:
            new_data[p] = "."
        elif data[k] == "." and v == 3:
            new_data[p] = "#"
        else:
            new_data[p] = "."


    data = copy.deepcopy(new_data)

t = 0
for k in data:
    if data[k] == "#":
        t += 1
print(t)

```
